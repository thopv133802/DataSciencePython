```python


import matplotlib.pyplot as pyplot

"""Setting styles"""

pyplot.style.use("classic")

"""How to display plots"""

"""Plotting from a script"""

%matplotlib inline

import numpy 

x = numpy.linspace(0, 10, 100)

figure = pyplot.figure()

pyplot.plot(x, numpy.sin(x), "-")
pyplot.plot(x, numpy.cos(x), "--")

figure.savefig("myFigure.png")
```

```python
"""Matlab stype interface"""

pyplot.figure()
pyplot.subplot(2, 1, 1)
pyplot.plot(x, numpy.sin(x))

pyplot.subplot(2, 1, 2)
pyplot.plot(x, numpy.cos(x))
```

```python

"""Object oriented interface"""

figure, axes = pyplot.subplots(2)

axes[0].plot(x, numpy.sin(x))
axes[1].plot(x, numpy.cos(x))
```

```python
"""Simple line plots"""

%matplotlib inline
import matplotlib.pyplot as pyplot

pyplot.style.use("seaborn-whitegrid")

import numpy 

figure = pyplot.figure()
axes = pyplot.axes()

x = numpy.linspace(0, 10, 1000)
axes.plot(x, numpy.sin(x))
```

```python
pyplot.plot(x, numpy.sin(x))
pyplot.plot(x, numpy.cos(x))
```

```python
"""Adjusting the Plot: Line colors and styles"""


"""Color"""
pyplot.plot(x, numpy.sin(x - 0), color = "blue")
pyplot.plot(x, numpy.sin(x - 1), color = "g")
pyplot.plot(x, numpy.sin(x - 2), color = "0.75")
pyplot.plot(x, numpy.sin(x - 3), color = "#FFDD44")
pyplot.plot(x, numpy.sin(x - 4), color = (1.0, 0.2, 0.3))
```

```python

"""Line style"""
pyplot.plot(x, x + 0, linestyle = "-")
pyplot.plot(x, x + 1, linestyle = "--")
pyplot.plot(x, x + 2, linestyle = "-.")
pyplot.plot(x, x + 3, linestyle = ":")
```

```python
"""Kt hp Color v Line Style"""
pyplot.plot(x, x + 0, "-g")
pyplot.plot(x, x + 1, "--c")
pyplot.plot(x, x + 2, "-.k")
pyplot.plot(x, x + 3, ":r")
```

```python
"""Axes limits"""

pyplot.plot(x, numpy.sin(x))

pyplot.xlim(-1, 11)
pyplot.ylim(-1.5, 1.5)
```

```python
pyplot.plot(x, numpy.sin(x))
pyplot.axis([-1, 11, -1.5, 1.5])
```

```python
"""m st cc trc"""

pyplot.plot(x, numpy.sin(x))
pyplot.axis([-1, 11, -1.5, 1.5])
pyplot.axis("tight")
```

```python
"""T l khung hnh bng nhau - mt n v x bng mt n v y"""
pyplot.plot(x, numpy.sin(x))
pyplot.axis("equal")
```

```python
"""Labeling plots"""

pyplot.plot(x, numpy.sin(x))
pyplot.title("A sine curve")
pyplot.xlabel("x")
pyplot.ylabel("sin(x)")
```

```python
pyplot.plot(x, numpy.sin(x), "-g", label = "sin(x)")
pyplot.plot(x, numpy.cos(x), ":b", label = "cos(x)")
pyplot.axis("equal")

pyplot.legend()
```

```python
axes = pyplot.axes()
axes.plot(x, numpy.sin(x))
axes.set(
    xlim = (0, 10),
    ylim = (-2, 2),
    xlabel = "x",
    ylabel = "sin(x)",
    title = "A simple plot"
)
axes.axis("equal")
```

```python
"""Simple scatter plots"""

%matplotlib inline
import matplotlib.pyplot as pyplot
pyplot.style.use("seaborn-whitegrid")
import numpy 

x = numpy.linspace(0, 10, 30)
y = numpy.sin(x)

pyplot.plot(x, y, "o", color = "black")
```

```python
"""Chuyn nghip hn vi .scatter() """

randomer = numpy.random.RandomState(0)
x = randomer.randn(100)
y = randomer.randn(100)

colors = randomer.rand(100)
sizes = 1000 * randomer.rand(100)

pyplot.scatter(x, y, c = colors, s = sizes, alpha = 0.3, cmap = "viridis")
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
pyplot.scatter(features[0], features[1], alpha = 0.2, s = 100 * features[3], c = iris.target, cmap = "viridis")
pyplot.xlabel(iris.features_names[0])
pyplot.ylabel(iris.features_names[1])
```
